```python
# 'teamgap' 열 추가 (차이가 0인 경우 0으로 처리)
final_ad_ap['teamgap'] = (final_ad_ap['team100_score'] > final_ad_ap['team200_score']).astype(int)
final_ad_ap.loc[final_ad_ap['team100_score'] == final_ad_ap['team200_score'], 'teamgap'] = 2  # 차이가 0인 경우 2로 설정

# 결과 출력
final_ad_ap
```

```python
# 각 팀의 골드 획득량 계산
team100_gold = final_ad_ap[['totalGold_100_{}'.format(i) for i in range(1, 6)]].sum(axis=1)
team200_gold = final_ad_ap[['totalGold_200_{}'.format(i) for i in range(6, 11)]].sum(axis=1)

# 새로운 열로 추가
final_ad_ap['team100gold'] = team100_gold
final_ad_ap['team200gold'] = team200_gold

final_ad_ap.iloc[:,-2:]
```

```python
# 각 팀 골드 차이 계산
final_ad_ap['goldGap'] = final_ad_ap['team100gold'] - final_ad_ap['team200gold']
# 'reverse' 파생 변수 생성
final_ad_ap['reverse'] = 0
final_ad_ap.loc[(final_ad_ap['winteam'] == '[100]') & (final_ad_ap['goldGap'] < 0), 'reverse'] = 1
final_ad_ap.loc[(final_ad_ap['winteam'] == '[200]') & (final_ad_ap['goldGap'] > 0), 'reverse'] = 1

final_ad_ap['reverse'].value_counts()
```

```python
# 'winteam' 열의 값이 [100]인 경우를 1로, [200]인 경우를 0으로 변경
final_ad_ap['winteam'] = final_ad_ap['winteam'].apply(lambda x: 1 if x == '[100]' else 0)

# 결과 출력
final_ad_ap
```

### teamgap에 따른 winteam 비율 확인
 - teamgap = 0 : tier score 가 200이 더 높은경우
 - teamgap = 1 : tier score 가 100이 더 높은경우
 - teamgap = 2 : 동일한 경우
#### 팀 100의 스코어가 더 높은경우는 팀 100의 승률이 더 높음
#### 팀 200의 스코어가 더 높은경우는 큰 차이를 보이지 않음
#### 팀 100, 200의 스코어가 동일한 경우는 팀 100의 스코어가 더 높음

```python
# teamgap과 winteam의 조합별 빈도수 계산
value_counts = df3.groupby(['teamgap', 'winteam']).size().unstack(fill_value=0)

# Seaborn을 사용하여 파이 차트 그리기
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, len(value_counts), figsize=(15, 5))

for idx, (gap, row) in enumerate(value_counts.iterrows()):
    sns.barplot(x=row.index, y=row.values, ax=axes[idx])
    axes[idx].set_title(f'Teamgap: {gap}')

plt.tight_layout()
plt.show()
```
