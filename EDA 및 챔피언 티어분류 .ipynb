### kda 변수 만들기

```python
# KDA 계산 함수 정의
def calculate_kda(kills, assists, deaths):
    if deaths == 0:
        return (kills + assists) * 1.2
    else:
        return (kills + assists) / deaths

# player1부터 player10까지의 KDA 계산 및 저장
for i in range(1, 11):
    final1[f'player{i}KDA'] = final1.apply(lambda row: calculate_kda(row[f'player{i}kills'], row[f'player{i}assists'], row[f'player{i}deaths']), axis=1)
```
