```python
import requests
import pandas as pd
import time
```

```python
match15 = pd.DataFrame() ;match15
```

```python
for i in range(len(matchids)):
    try:
        matchtime = 'https://asia.api.riotgames.com/lol/match/v5/matches/' + matchids[i] + '/timeline?api_key=' + api_key 
        r = requests.get(matchtime)
        match = 'https://asia.api.riotgames.com/lol/match/v5/matches/' + matchids[i] + '?api_key=' + api_key
        j = requests.get(match)
    
        while r.status_code == 429:
            time.sleep(5)
            matchtime = 'https://asia.api.riotgames.com/lol/match/v5/matches/' + matchids[i] + '/timeline?api_key=' + api_key 
            r = requests.get(matchtime)
            match = 'https://asia.api.riotgames.com/lol/match/v5/matches/' + matchids[i] + '?api_key=' + api_key
            j = requests.get(match)
        
        df_part = pd.DataFrame(j.json())
        participants_data = df_part['info']['participants']
        participants_df = pd.DataFrame(participants_data)
        df_time = pd.DataFrame(r.json())
        df_time = pd.DataFrame(df_time['info']['frames'])
        df_time15 = pd.DataFrame(df_time['participantFrames'][15])
    
       
        match15.loc[i,'matchid'] = matchids[i]
    
    ###############################################
    

        # match15 데이터프레임에 필요한 정보 추가
        # 챔피언 이름 할당
        for j in range(1, 6):
            match15.loc[i,f'championName_100_{j}'] = participants_df.loc[j-1, 'championName']
        for j in range(6, 11):
            match15.loc[i,f'championName_200_{j}'] = participants_df.loc[j-1, 'championName']
        #팀원 포지션 할당
        for j in range(1, 6):
            match15.loc[i,f'teamPosition_100_{j}'] = participants_df.loc[j-1, 'teamPosition']
        for j in range(6, 11):
            match15.loc[i,f'teamPosition_200_{j}'] = participants_df.loc[j-1, 'teamPosition']
        #소환사 별 15분 골드 할당
        for j in range(1, 6):
            match15.loc[i,f'totalGold_100_{j}'] = df_time15.iloc[10, j-1]
        for j in range(6, 11):
            match15.loc[i,f'totalGold_200_{j}'] = df_time15.iloc[10, j-1]
        #소환사 별 15분 레벨 할당 
        for j in range(1, 6):
            match15.loc[i,f'level_100_{j}'] = df_time15.iloc[5, j-1]
        for j in range(6, 11):
            match15.loc[i,f'level_200_{j}'] = df_time15.iloc[5, j-1]
        #소환사 별 15분 미니언 처치수 할당
        for j in range(1, 6):
            match15.loc[i,f'minionsKilled_100_{j}'] = df_time15.iloc[6, j-1]
        for j in range(6, 11):
            match15.loc[i,f'minionsKilled_200_{j}'] = df_time15.iloc[6, j-1]
        #소환사 별 15분 정글 미니언 처치수 할당 
        for j in range(1, 6):
            match15.loc[i,f'jungleMinionsKilled_100_{j}'] = df_time15.iloc[4, j-1]
        for j in range(6, 11):
            match15.loc[i,f'jungleMinionsKilled_200_{j}'] = df_time15.iloc[4, j-1]
             
        print(i, 'done')
        
    except Exception as e:
        print(i, f"Error occurred for matchid {matchids[i]}: {e}")
        pass
```
