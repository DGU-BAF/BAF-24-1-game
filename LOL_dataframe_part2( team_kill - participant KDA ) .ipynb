## df_team_kill: 블루팀과 레드팀의 퍼스트 블러드 여부
### 팀별 KDA

```python
# 변수 초기화
blue_kill, red_kill = 0, 0
blue_firstkill, red_firstkill = 0, 0
blue_assist, red_assist = 0, 0
red_death, blue_death = 0, 0

for i in range(0, 16):
    # DataFrame으로 변환합니다.
    df = pd.DataFrame(df3['events'][i])
    # 'type' 열이 'CHAMPION_KILL'인 행들을 선택합니다.
    selected_champion_kills = df[df['type'] == 'CHAMPION_KILL']
    
    for index, event in selected_champion_kills.iterrows():
        # 킬러의 팀을 확인합니다.
        if event['killerId'] in range(1, 6):  # Blue Team
            if red_kill == 0 and blue_kill == 0:
                blue_firstkill += 1
            blue_kill += 1
            assist_ids = event['assistingParticipantIds']
            if isinstance(assist_ids, list):
                blue_assist += len(assist_ids)
            red_death += 1
        elif event['killerId'] in range(6, 11):  # Red Team
            if red_kill == 0 and blue_kill == 0:
                red_firstkill += 1
            red_kill += 1
            assist_ids = event['assistingParticipantIds']
            if isinstance(assist_ids, list):
                red_assist += len(assist_ids)
            blue_death += 1
            

# 결과 출력
print("Blue Team:")
print("Total Kills:", blue_kill)
print("First Kills:", blue_firstkill)
print("Total Assists:", blue_assist)
print("Red Team:")
print("Total Kills:", red_kill)
print("First Kills:", red_firstkill)
print("Total Assists:", red_assist)

# 각 팀의 킬 수를 담은 리스트 생성
blue_firstkill_list = [blue_firstkill]
red_firstkill_list = [red_firstkill]

# 데이터프레임 생성
df_team_kill = pd.DataFrame(data={'blue_firstkill': blue_firstkill_list, 'red_firstkill': red_firstkill_list})
df_team_kill
```

## df_monster:드래곤, 퍼스트 드래곤, 호드 처치 수

```python
# 변수 초기화
blue_elite, red_elite = 0, 0
blue_HORDE, red_HORDE = 0, 0
blue_dragon, red_dragon = 0, 0
blue_baron, red_baron = 0, 0
blue_firstdragon, red_firstdragon = 0, 0
blue_dragontype, red_dragontype = [], []

for i in range(0, 16):
    # DataFrame으로 변환합니다.
    df = pd.DataFrame(df3['events'][i])
    # 'type' 열이 'ELITE_MONSTER_KILL'인 행들을 선택합니다.
    selected_ELITE_MONSTER_KILL = df[df['type'] == 'ELITE_MONSTER_KILL']
    
    for index, event in selected_ELITE_MONSTER_KILL.iterrows():
        # 이벤트의 유형을 확인합니다.
        if event['monsterType'] == 'DRAGON':
            # 킬러의 팀을 확인합니다.
            if event['killerId'] in range(1, 6):  # Blue Team
                blue_dragon += 1
                if blue_dragon == 1 and red_dragon == 0:
                    blue_firstdragon += 1
                blue_dragontype.append(event['monsterSubType'])
            else:  # Red Team
                red_dragon += 1
                if red_dragon == 1 and blue_dragon == 0:
                    red_firstdragon += 1
                red_dragontype.append(event['monsterSubType'])
        elif event['monsterType'] == 'HORDE':
            if event['killerId'] in range(1, 6):  # Blue Team
                blue_HORDE += 1
            else:  # Red Team
                red_HORDE += 1
        elif event['monsterType'] == 'BARON_NASHOR':
            if event['killerId'] in range(1, 6):  # Blue Team
                blue_baron += 1
                if blue_baron == 1 and red_baron == 0:
                    blue_firstbaron += 1
            else:  # Red Team
                red_baron += 1
                if red_baron == 1 and blue_baron == 0:
                    red_firstbaron += 1
# 결과 출력
print("Blue Team:")
print("Total Dragon Kills:", blue_dragon)
print("First Dragon:", blue_firstdragon)
print("Dragon Types:", blue_dragontype)
print("Total Rift Herald Kills:", blue_HORDE)
print("Total Baron Kills:", blue_baron)
print("First Baron:", blue_firstbaron)

print("\nRed Team:")
print("Total Dragon Kills:", red_dragon)
print("First Dragon:", red_firstdragon)
print("Dragon Types:", red_dragontype)
print("Total Rift Herald Kills:", red_HORDE)
print("Total Baron Kills:", red_baron)
print("First Baron:", red_firstbaron)

```

## df_participant_KDA: 개인별 KDA 

```python
# 변수 초기화
participant_stats = {}

for i in range(0, 16):
    # DataFrame으로 변환합니다.
    df = pd.DataFrame(df3['events'][i])
    # 'type' 열이 'CHAMPION_KILL'인 행들을 선택합니다.
    selected_champion_kills = df[df['type'] == 'CHAMPION_KILL']
    
    for index, event in selected_champion_kills.iterrows():
        killer_id = event['killerId']
        if killer_id not in participant_stats:
            participant_stats[killer_id] = {'kills': 0, 'assists': 0, 'deaths': 0}
        
        participant_stats[killer_id]['kills'] += 1
        
        assist_ids = event['assistingParticipantIds']
        if isinstance(assist_ids, list):
            participant_stats[killer_id]['assists'] += len(assist_ids)
        
        # 사망자도 킬 수에 포함되므로 사망 수도 증가시킵니다.
        death_ids = event['victimId']
        if death_ids not in participant_stats:
            participant_stats[death_ids] = {'kills': 0, 'assists': 0, 'deaths': 0}
        
        participant_stats[death_ids]['deaths'] += 1

# 결과 출력
for participant_id, stats in participant_stats.items():
    print(f"Participant {participant_id}: Kills={stats['kills']}, Assists={stats['assists']}, Deaths={stats['deaths']}")
```

```python
# 데이터 프레임 안에 넣기 
participant_kills = {}
participant_assists = {}
participant_deaths={}


for participant_id, stats in participant_stats.items():
    participant_kills[f"participant{participant_id}_kill"] = stats['kills']
    participant_assists[f"participant{participant_id}_ass"] = stats['assists']
    participant_deaths[f"participant{participant_id}_deaths"] = stats['deaths']
    
df_kill = pd.DataFrame.from_dict(participant_kills, orient='index').T
df_assist=pd.DataFrame.from_dict(participant_assists, orient='index').T
df_deaths=pd.DataFrame.from_dict(participant_deaths, orient='index').T    


# 수평으로 데이터프레임 결합
df_participant_KDA = pd.concat([df_kill, df_assist, df_deaths], axis=1)
df_participant_KDA
```
